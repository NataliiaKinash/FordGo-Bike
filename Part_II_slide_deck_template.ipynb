## Investigation Overview


The key goal of the presentation is to demonstrate the relation between the number of rides and subscriber type depending on the day of the week and the hour of the day. 

Analysis of ride duration depending on the age and gender is also included into the presentation.

## Dataset Overview

This data set includes information about individual rides made in a bike-sharing system covering the greater San Francisco Bay area.

```python
# import all packages and set plots to be embedded inline
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sb

%matplotlib inline

# suppress warnings from final output
import warnings
warnings.simplefilter("ignore")
```

```python
# load in the dataset into a pandas dataframe
bike = pd.read_csv('bike.csv')
```

## Bike Usage per Day


Customers are using bikes more or less uniformly during the week, while there is significant reduction of bike usege during the weekends among subscribers.

```python
day_name = ["Monday","Tuesday","Wednesday","Thursday","Friday","Saturday","Sunday"]

base_color = sb.color_palette()[0]
plt.figure(figsize=(10,6))
plt.title('Bike Usage per Day', fontsize=15)
chart = sb.countplot(data=bike, x='day', order=day_name, hue='user_type')
```

## Hourly Usage per Day

Though there are slight spikes during the rush hours the behaviour of customers is again mostly uniform during the day, whereas subscribers are using the bike a lot more actively during the rush hours.

```python
hours = np.arange(0,24)

base_color = sb.color_palette()[0]
plt.figure(figsize=(10,6))
plt.title('Hourly Usage per Day', fontsize=15)
chart = sb.countplot(data=bike, x='hour', hue='user_type')
```

## Ride Duration for Different Age and Gender


Most of the users do not use the bike for more than 5 hours. People, who have not identified their gender mostly make short-time rides. In general there are more service users among younger people and short rides are more popular than the long rides. Though the amount of Female users in significantly smaller than the amount of Male users their behaviour as users does not differ much.


```python
plt.figure(figsize = (15,7))
ax = sb.stripplot(x="age_bin", y="duration_hours", hue="member_gender", data=bike, jitter=0.4, alpha=0.5).set_title("Ride Duration for Different Age Group and Gender")
```
